```python
import pandas as pd
import matplotlib.pyplot as plt
import random

# Membaca data dari file CSV
data = pd.read_csv('/content/diamonds.csv')

# Daftar fitur kategorikal
categorical_features = ['cut', 'color', 'clarity']

# Membuat bar chart untuk setiap fitur kategorikal dengan warna yang berbeda
for feature in categorical_features:
    category_counts = data[feature].value_counts()

    # Membuat bar chart dengan warna yang berbeda pada setiap sumbu y
    plt.figure(figsize=(8, 6))
    colors = [plt.cm.jet(random.random()) for _ in range(len(category_counts))]
    category_counts.plot(kind='bar', color=colors)
    plt.title(f'Bar Chart dari Fitur Kategorikal: {feature}')
    plt.xlabel(feature)
    plt.ylabel('Jumlah')
    plt.xticks(rotation=45)  # Untuk memutar label sumbu x jika diperlukan
    plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Membaca data dari file CSV
data = pd.read_csv('/content/diamonds.csv')

# Daftar fitur kategorikal
categorical_features = ['cut', 'color', 'clarity']

# Membuat pie chart untuk setiap fitur kategorikal dalam setiap fitur
for feature in data.columns:
    if feature in categorical_features:
        # Menghitung jumlah masing-masing kategori dalam fitur kategorikal
        category_counts = data[feature].value_counts()

        # Membuat pie chart
        plt.figure(figsize=(8, 6))
        plt.pie(category_counts, labels=category_counts.index, autopct='%1.1f%%', startangle=90)
        plt.title(f'Pie Chart dari Fitur Kategorikal: {feature}')

        # Menampilkan pie chart
        plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Membaca data dari file CSV
data = pd.read_csv('/content/diamonds.csv')

# Daftar fitur numerik
numerical_features = ['carat', 'depth', 'table', 'x', 'y', 'z']

# Mengatur susunan subplot
fig, axes = plt.subplots(2, 3, figsize=(15, 10))

# Membuat histogram distribusi untuk setiap fitur numerik
for i, feature in enumerate(numerical_features):
    row = i // 3
    col = i % 3
    ax = axes[row, col]

    data[feature].plot(kind='hist', bins=30, edgecolor='k', ax=ax)
    ax.set_title(f'Histogram Distribusi {feature}')
    ax.set_xlabel(feature)
    ax.set_ylabel('Jumlah')

# Mengatur tata letak subplot
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import plotly.express as px
import ipywidgets as widgets
from ipywidgets import interactive

# Membaca data dari file CSV
data = pd.read_csv('/content/diamonds.csv')

# Daftar fitur numerik
numerical_features = ['carat', 'depth', 'table', 'x', 'y', 'z']

def plot_histogram(bin_width):
    fig = px.histogram(data, x=numerical_features, nbins=bin_width, facet_col='variable')
    fig.update_layout(
        title=f'Histogram Distribusi - Bin Width: {bin_width}',
        xaxis_title='Nilai',
        yaxis_title='Jumlah'
    )
    fig.show()

# Membuat widget untuk mengontrol bin width
bin_width_slider = widgets.IntSlider(value=20, min=5, max=100, step=5, description='Bin Width')

# Membuat tampilan interaktif
interactive_plot = interactive(plot_histogram, bin_width=bin_width_slider)
output = interactive_plot.children[-1]
output.layout.height = '400px'
interactive_plot
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Membaca data dari file CSV
data = pd.read_csv('/content/diamonds.csv')

# Daftar fitur numerik
numerical_features = ['carat', 'depth', 'table', 'x', 'y', 'z']

# Daftar bin width yang akan diuji
bin_widths = [5, 10, 20, 30]

# Membuat grafik histogram untuk tiap fitur dengan berbagai bin width
for feature in numerical_features:
    plt.figure(figsize=(15, 6))
    plt.suptitle(f'Histogram Distribusi {feature}', fontsize=16)

    for i, bin_width in enumerate(bin_widths):
        plt.subplot(1, len(bin_widths), i + 1)
        plt.hist(data[feature], bins=range(int(data[feature].min()), int(data[feature].max()) + bin_width, bin_width), edgecolor='k')
        plt.title(f'Bin Width = {bin_width}')
        plt.xlabel(feature)
        plt.ylabel('Jumlah')

    plt.tight_layout(rect=[0, 0, 1, 0.92])
    plt.show()
```
